```python
import pandas as pd
import numpy as np

df = pd.read_csv("process_counts300.csv")
```

```python
for i in range(30):
    if i != 0:
        df.iloc[i] = df.iloc[i] + df.iloc[i-1]
```

```python
df["block"] = np.linspace(10000, 300000, num=len(df))
```

```python
processes = []
for c in df.columns:
    if c != "block":
        processes.append(c)

```

```python
df[processes] = df[processes].div(df["block"], axis=0)
```

```python
import matplotlib.pyplot as plt

for process in processes:
    plt.plot(df["block"], df[process], label=process)
    plt.axhline(y=df[process].iloc[-1] * 0.98, color='r', linestyle='--', label=f"mean + 2%")
    plt.axhline(y=df[process].iloc[-1] * 1.02, color='g', linestyle='--', label=f"mean - 2%")
    plt.xlabel("Simulations")
    plt.ylabel("Count/Sim")
    plt.title(f"{process} Counts/Simulation")
    plt.legend()
    plt.show()
```
